```python
import numpy as np
import matplotlib.pyplot as plt

k   = 0.0016   # transconductance (already calculated)
Vth = 0.7      # threshold voltage (V)

print("k =", k)
print("Vth =", Vth)
```

```python
# Gate voltage list from 0 to 3V
Vgs = np.linspace(0, 3, 100)

# Calculate current for each voltage
Id = []
for v in Vgs:
    if v < Vth:
        Id.append(0)
    else:
        Id.append(0.5 * k * (v - Vth)**2 * 1000)

# Plot
plt.plot(Vgs, Id)
plt.xlabel('Gate Voltage (V)')
plt.ylabel('Drain Current (mA)')
plt.title('Transfer Curve')
plt.grid(True)
plt.show()
```

```python
Vds = np.linspace(0, 3, 100)

for Vgs_fixed in [1.0, 1.5, 2.0, 2.5, 3.0]:
    Id = []
    for v in Vds:
        if Vgs_fixed < Vth:
            Id.append(0)
        elif v < (Vgs_fixed - Vth):
            Id.append(k * ((Vgs_fixed - Vth)*v - 0.5*v**2) * 1000)
        else:
            Id.append(0.5 * k * (Vgs_fixed - Vth)**2 * 1000)

    plt.plot(Vds, Id, label='Vgs = ' + str(Vgs_fixed) + 'V')

plt.xlabel('Drain Voltage (V)')
plt.ylabel('Drain Current (mA)')
plt.title('Output Curves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Vgs = np.linspace(0, 3, 100)

for L in [2, 1, 0.5, 0.25]:
    k_new = 0.0016 * (1 / L)   # k doubles as L halves
    Id = []
    for v in Vgs:
        if v < Vth:
            Id.append(0)
        else:
            Id.append(0.5 * k_new * (v - Vth)**2 * 1000)

    plt.plot(Vgs, Id, label='L = ' + str(L) + 'um')

plt.xlabel('Gate Voltage (V)')
plt.ylabel('Drain Current (mA)')
plt.title('Channel Length Scaling')
plt.legend()
plt.grid(True)
plt.savefig('channel_scaling.png', dpi=200)
plt.show()
```
